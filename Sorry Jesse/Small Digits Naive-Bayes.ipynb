```python
all_data_in = np.genfromtxt('smalldigits.csv', delimiter=',')
all_data = all_data_in[:1438]
all_data_test = all_data_in[1438:]
#unique outcomes and their count
uniqueLabels, labelCounts = np.unique(all_data[:,64],return_counts = True)
total_datapoints = np.sum(labelCounts)
print(all_data)
#print(all_data.shape)
print(all_data_test)
#print(all_data_test.shape)
print(uniqueLabels,labelCounts)
#print(np.sum(labelCounts))
```

Declare the main table

```python
main_table = np.zeros(shape = (64,10))
#print(main_table)
```

Fill in the main table

```python
for labelNum in range(10):
    data_for_curr_label = all_data[all_data[:,64] == labelNum]
    denominator = data_for_curr_label.shape[0]
    print(denominator)
    #For each pixel get prob of pixel being on
    for pixelInd in range(64):
        #count how many times 1 occurs in col:pixelInd from data_for_curr_label
        #print(data_for_curr_label[:,5])
        unique,counts = np.unique(data_for_curr_label[:,pixelInd] ,return_counts = True)
        
        #calculate the probability
        tempIndex = 0
        numerator = 0
        for pixelValue in unique:
            if pixelValue == 1:
                numerator = counts[tempIndex]
                break
            tempIndex = tempIndex+1
            
        main_table[pixelInd][labelNum] = float(numerator/denominator)
        

print(main_table)
print(main_table.shape)
```

Function to predict data:

```python
def predict_value(data_point):
    #data_point is vector(64) with pixels
    numerator_values = np.zeros(10)
    denominator = 0.0
    for i in range(10):
        curr_total = 1.0
        for j in range(len(data_point)):
            currPixel = data_point[j]
            if currPixel == 0:
                #pixel is off
                curr_total = curr_total*(1-main_table[j][i])
            else:
                curr_total = curr_total*(main_table[j][i])
                
        numerator_values[i] = curr_total*(labelCounts[i]/total_datapoints)
        denominator = denominator + numerator_values[i]
    
    #For each label get the prob
    largestProb = -1
    largestProbLabelNum = -1
    for labelNum in range(10):
        curr_prob = numerator_values[labelNum]/denominator
        
        #print(labelNum,curr_prob)
        
        if curr_prob > largestProb:
            largestProb = curr_prob
            largestProbLabelNum = labelNum
        
    
    return largestProbLabelNum
```

Function to test data:

```python
total_num_tests = len(all_data_test)
total_num_correct = 0
#Predict each data_point
for i in range(len(all_data_test)):
    test_dataPoint = all_data_test[i][:64]
    test_dataPointAnswer = all_data_test[i][64]
    test_predict_answer = predict_value(test_dataPoint)
    if test_dataPointAnswer == test_predict_answer:
        total_num_correct = total_num_correct + 1

accuracy = total_num_correct/total_num_tests
print(accuracy)
```
